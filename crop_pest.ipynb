```python
import os
base_dir = 'cashew/Cashew_crop'
print("contents of base directory:")
print(os.listdir(base_dir))
print("contents of test directory:")
print(os.listdir(f'{base_dir}/test'))
print("contents of train directory:")
print(os.listdir(f'{base_dir}/train'))
```

```python
import os
train_dir = os.path.join(base_dir,'train')
test_dir = os.path.join(base_dir,'test')

#training directories
train_antracnose_dir = os.path.join(train_dir,'Cashew_anthracnose_train')
train_gumosis_dir = os.path.join(train_dir,'Cashew_gumosis_train')
train_healthy_dir = os.path.join(train_dir,'Cashew_healthy_train')
train_leafminer_dir = os.path.join(train_dir,'Cashew_leaf_miner_train')
train_redrust_dir = os.path.join(train_dir,'Cashew_red_rust_train')


#testing directories
test_antracnose_dir = os.path.join(test_dir,'Cashew_anthracnose_test')
test_gumosis_dir = os.path.join(test_dir,'Cashew_gumosis_test')
test_healthy_dir = os.path.join(test_dir,'Cashew_healthy_test')
test_leafminer_dir = os.path.join(test_dir,'Cashew_leaf_miner_test')
test_redrust_dir = os.path.join(test_dir,'Cashew_red_rust_test')
```

```python
print('total training anthracnose images :', len(os.listdir(    train_antracnose_dir  ) ))
print('total training gumosis images :', len(os.listdir(       train_gumosis_dir) ))
print('total training healthy images :', len(os.listdir(     train_healthy_dir) ))
print('total training leaf miner images :', len(os.listdir(       train_leafminer_dir) ))
print('total training red rust images :', len(os.listdir(       train_redrust_dir) ))

print('total test anthracnose images :', len(os.listdir( test_antracnose_dir ) ))
print('total test gumosis images :', len(os.listdir( test_gumosis_dir ) ))
print('total test healthy images :', len(os.listdir( test_healthy_dir ) ))
print('total test leaf miner images :', len(os.listdir( test_leafminer_dir ) ))
print('total test red rust images :', len(os.listdir( test_redrust_dir ) ))
```

```python
import tensorflow as tf

model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(16, (3,3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(5, activation='sigmoid')
])
```

```python
model.summary()
```

```python
from tensorflow.keras.optimizers import RMSprop

model.compile(optimizer=RMSprop(learning_rate=0.001),
              loss='categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1.0/255.)
test_datagen = ImageDataGenerator(rescale=1.0/255.)

train_generator = train_datagen.flow_from_directory(train_dir,
                                                    batch_size=20,
                                                    class_mode='categorical',
                                                    target_size=(150, 150))

test_generator =  test_datagen.flow_from_directory(test_dir,
                                                   batch_size=20,
                                                   class_mode='categorical',
                                                   target_size=(150, 150))
```

```python
history = model.fit(
            train_generator,
            epochs=15,
            validation_data=test_generator,
            verbose=2
            )
```
